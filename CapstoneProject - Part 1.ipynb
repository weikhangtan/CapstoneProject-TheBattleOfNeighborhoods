## 1. Data
### Download and Explore Dataset

I will use the exisiting free New York dataset in the link: https://geo.nyu.edu/catalog/nyu_2451_34572 to determine the number of borough and neighborhoods. In order to segment the neighborhoods and explore them, we will essentially need a dataset that contains the 5 boroughs and the neighborhoods that exist in each borough as well as the the latitude and logitude coordinates of each neighborhood. 


For convenience, I have downloaded the dataset and save to my directory via this link https://cocl.us/new_york_dataset

#### Load and explore the data

Load the data in json file

```python
with open('newyork_data.json') as json_data:
    newyork_data = json.load(json_data)
```

Quick look on the dataset

```python
newyork_data
```

Define a variable for the `features` key, which is basically list of neighborhoods

```python
neighborhoods_data = newyork_data['features']
#Quick look on the first neighborhood data
neighborhoods_data[0]
```

#### Tranform the data into a *pandas* dataframe

```python
# define the dataframe columns
column_names = ['Borough', 'Neighborhood', 'Latitude', 'Longitude'] 

# instantiate the dataframe
neighborhoods = pd.DataFrame(columns=column_names)
neighborhoods
```

Loop in the neighborhood data into the dataframe for each rows

```python
for data in neighborhoods_data:
    borough = neighborhood_name = data['properties']['borough'] 
    neighborhood_name = data['properties']['name']
        
    neighborhood_latlon = data['geometry']['coordinates']
    neighborhood_lat = neighborhood_latlon[1]
    neighborhood_lon = neighborhood_latlon[0]
    
    neighborhoods = neighborhoods.append({'Borough': borough,
                                          'Neighborhood': neighborhood_name,
                                          'Latitude': neighborhood_lat,
                                          'Longitude': neighborhood_lon}, ignore_index=True)
```

```python
#Quick look the neighborhood dataframe
neighborhoods
```

Check the number of borough and neighborhoods in New York

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(neighborhoods['Borough'].unique()),
        neighborhoods.shape[0]
    )
)
```

#### Use geopy to get the latitude and longitude values of New York.

```python
address = 'New York City, NY'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of New York City are {}, {}.'.format(latitude, longitude))
```

#### Create the map for New York city and mark down each of the neighborhoods

```python
# create map of New York using latitude and longitude values
map_newyork = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```

#### Find the borough with the most neighborhood

```python
neighborhoods.groupby('Borough').count()
```

From the above table, we can see `Queens` borough has the highest number of neighborhood, hence I will focus and pick up `Queens` borough for my further segmenting and clustering.

Now, create a new DataFrame to capture the Queens's neighborhoods

```python
queens_data = neighborhoods[neighborhoods['Borough'] == 'Queens'].reset_index(drop=True)
queens_data
```

Let's get the geographical coordinates of Queens via geopy.

```python
address = 'Queens, NY'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
queens_latitude = location.latitude
queens_longitude = location.longitude
print('The geograpical coordinate of Manhattan are {}, {}.'.format(queens_latitude, queens_longitude))
```

Visualize the Queens Neighborhoods in the map

```python
# create map of Queens using latitude and longitude values
map_queens = folium.Map(location=[queens_latitude, queens_longitude], zoom_start=10)

# add markers to map
for lat, lng, label in zip(queens_data['Latitude'], queens_data['Longitude'], queens_data['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_queens)  
    
map_queens
```
